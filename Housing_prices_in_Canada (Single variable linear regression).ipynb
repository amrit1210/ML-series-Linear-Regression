```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"data/canada_per_capita_income.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
%matplotlib inline
plt.scatter(df.year, df[['per capita income (US$)']], color='red', marker='x')
plt.xlabel("Year")
plt.ylabel("per capita income (US$)")
plt.title("Housing prices in Canada (yearly)")
```

```python
reg = linear_model.LinearRegression()
```

```python
reg
```

```python
reg.fit(X=df[['year']], y=df['per capita income (US$)'])
```

```python
%matplotlib inline
plt.scatter(df.year, df[['per capita income (US$)']], color='red', marker='x')
plt.xlabel("Year")
plt.ylabel("per capita income (US$)")
plt.title("Housing prices in Canada (yearly)")
plt.plot(df[['year']], reg.predict(df[['year']]), color='blue') 
```

```python
#Since y=mx+b, m - slope(gradient), b= intercept
m = reg.coef_
b = reg.intercept_
print(f"Slope(gradient) m : {m}")
print(f"Intercept b: {b}")
```
